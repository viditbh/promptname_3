```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plb
```

```python
df=pd.read_csv(r'C:\Users\Hp\OneDrive\Documents\airbnb (1).csv')
```

```python
df.head(5)
```

```python
df["city"]=df["address"].str.split(',').str[0]
df
```

```python
citi=['New Delhi','Mumbai']
county='Turkey'
def loca(x):
    if x ['city']== 'New Delhi':
        return 'New Delhi'
    elif x ['city']== 'Mumbai':
        return 'Mumbai'
    elif 'Turkey'in x['address']:
        return 'Turkey'
    else:
         return "other"
        

```

```python
df['category']=df.apply(loca,axis=1)
df
```

```python
y=df['category'].value_counts(normalize=True)*(100)
```

```python
plb.figure(figsize=(10,5))
y.plot(kind='pie',autopct='%1.1f%%')
plb.title("Location of Airbnbs")
plb.show()
```

```python
df['rate']=pd.to_numeric(df['rating'],errors='coerce')
df
```

```python
hi_count=df[df['rate']>=4].shape[0]
hi_count
low_count=df[df['rate']<4].shape[0]
```

```python
plb.figure(figsize=(10,6))
plb.pie([hi_count,low_count],labels=['More than 4 star rating', 'other'],autopct='%1.1f%%')
plb.title('Best airbnbs')
plb.show()
```
